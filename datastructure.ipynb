## 17. Describe the time complexity of accessing elements in a dictionary.

####-->The time complexity of accessing elements in a dictionary in Python is generally O(1)(constant time).


*   This is because dictionaries use a hash table internally,which allows direct access to values using keys without scanning all elements.
*   However,in rare cases of hash collisions,access time can degrade,but python handles collisions efficiently,so average access remains O(1)
